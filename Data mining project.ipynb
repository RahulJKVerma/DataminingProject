Function for converting string such as "5.0 star rating" and "5 friends" to numerical values using regex

Finding the count of Adverb and Adjective/Adverb combinations like "truly beautiful" or "exceptionally good" using nltk tokens and english corpus

```python
def adv_adj(sentence):
    count = 0
    text = nltk.word_tokenize(sentence)
    tagged = nltk.pos_tag(text)
    for (w1,t1), (w2,t2) in nltk.bigrams(tagged):
        if (t1.startswith('RB') and t2.startswith('JJ')) or (t1.startswith('RB') and t2.startswith('JJ')):
            count+=1
    return count/float(len(sentence))
```

Finding the count of Adjetives combinations like "wonderful" or "nice" using again nltk tokens and english corpus.

```python
def adj(sentence):
    count = 0
    text = nltk.word_tokenize(sentence)
    tagged = nltk.pos_tag(text)
    for (w1,t1) in tagged:
        if (t1.startswith('JJ')):
            count+=1
    return count/float(len(sentence))
```

Finding the count of personal pronouns like "Jack's" or "John" using nltk tokens and english corpus. This is differentiated from general nouns such as car and wife.

```python
def pers_noun(sentence):
    count = 0
    text = nltk.word_tokenize(sentence)
    tagged = nltk.pos_tag(text)
    for (w1,t1) in tagged:
        if (t1.startswith('NNP')):
            count+=1
    return count/float(len(sentence))
```
